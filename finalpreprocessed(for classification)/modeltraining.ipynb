```python
# Lightning.ai Ensemble Model for Candlestick Pattern Classification
# Dataset: Canny edge-detected candlestick images (Buy/Sell signals)
# Models: EfficientNetB0 + Vision Transformer Ensemble

import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

# ==================== CONFIGURATION ====================
DATASET_PATH = '/teamspace/studios/this_studio/finalpreprocessed'
MODEL_SAVE_PATH = '/teamspace/studios/this_studio/ensemble_final_model.h5'
CHECKPOINT_DIR = '/teamspace/studios/this_studio/checkpoints'

# Training parameters
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 1e-4
VALIDATION_SPLIT = 0.2
TARGET_ACCURACY = 0.95

# ViT parameters
PATCH_SIZE = 16
NUM_PATCHES = (IMG_SIZE // PATCH_SIZE) ** 2
PROJECTION_DIM = 512
NUM_HEADS = 8
TRANSFORMER_LAYERS = 6

# Class names
CLASSES = ['buyinput', 'sellinput']
NUM_CLASSES = 2

print("=" * 70)
print("ENSEMBLE CANDLESTICK PATTERN CLASSIFIER")
print("EfficientNetB0 + Vision Transformer")
print("=" * 70)
print(f"Dataset Path: {DATASET_PATH}")
print(f"Image Size: {IMG_SIZE}x{IMG_SIZE}")
print(f"Batch Size: {BATCH_SIZE}")
print(f"Epochs: {EPOCHS}")
print(f"Target Accuracy: {TARGET_ACCURACY * 100}%")
print("=" * 70)

# ==================== CREATE DIRECTORIES ====================
os.makedirs(CHECKPOINT_DIR, exist_ok=True)
print(f"âœ“ Checkpoint directory: {CHECKPOINT_DIR}")

# ==================== DATASET VALIDATION ====================
print("\nValidating dataset and checking for corrupted images...")

total_valid = 0
total_corrupted = 0
corrupted_files = []

for class_name in CLASSES:
    class_path = os.path.join(DATASET_PATH, class_name)
    if not os.path.exists(class_path):
        print(f"âœ— ERROR: {class_name} directory not found!")
        continue
    
    valid_count = 0
    corrupt_count = 0
    
    for filename in os.listdir(class_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
            file_path = os.path.join(class_path, filename)
            try:
                img = Image.open(file_path)
                img.verify()
                img.close()
                img = Image.open(file_path)
                img.load()
                img.close()
                valid_count += 1
            except Exception as e:
                corrupt_count += 1
                corrupted_files.append(file_path)
    
    total_valid += valid_count
    total_corrupted += corrupt_count
    print(f"  {class_name}: {valid_count} valid, {corrupt_count} corrupted")

print(f"\nâœ“ Total valid images: {total_valid}")
if total_corrupted > 0:
    print(f"âš  Found {total_corrupted} corrupted images - removing them...")
    for file_path in corrupted_files:
        try:
            os.remove(file_path)
            print(f"  Removed: {os.path.basename(file_path)}")
        except:
            pass
    print(f"âœ“ Cleaned up {len(corrupted_files)} corrupted files")

# ==================== DATA GENERATORS ====================
print("\nInitializing data generators (NO augmentation)...")

# Only rescaling - no augmentation for edge-detected images
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=VALIDATION_SPLIT
)

train_generator = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=42,
    color_mode='rgb'
)

validation_generator = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=42,
    color_mode='rgb'
)

print(f"âœ“ Training samples: {train_generator.samples}")
print(f"âœ“ Validation samples: {validation_generator.samples}")
print(f"âœ“ Class mapping: {train_generator.class_indices}")

# ==================== BUILD VISION TRANSFORMER ====================
print("\nBuilding Vision Transformer model...")

class Patches(layers.Layer):
    def __init__(self, patch_size):
        super().__init__()
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim):
        super().__init__()
        self.num_patches = num_patches
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded

def create_vit_model():
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    
    patches = Patches(PATCH_SIZE)(inputs)
    encoded_patches = PatchEncoder(NUM_PATCHES, PROJECTION_DIM)(patches)
    
    for _ in range(TRANSFORMER_LAYERS):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        attention_output = layers.MultiHeadAttention(
            num_heads=NUM_HEADS, key_dim=PROJECTION_DIM // NUM_HEADS, dropout=0.1
        )(x1, x1)
        x2 = layers.Add()([attention_output, encoded_patches])
        x3 = layers.LayerNormalization(epsilon=1e-6)(x2)
        x3 = layers.Dense(PROJECTION_DIM * 2, activation=tf.nn.gelu)(x3)
        x3 = layers.Dropout(0.1)(x3)
        x3 = layers.Dense(PROJECTION_DIM)(x3)
        x3 = layers.Dropout(0.1)(x3)
        encoded_patches = layers.Add()([x3, x2])
    
    representation = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    representation = layers.GlobalAveragePooling1D()(representation)
    representation = layers.Dropout(0.3)(representation)
    
    vit_output = layers.Dense(256, activation='relu', name='vit_features')(representation)
    
    model = keras.Model(inputs=inputs, outputs=vit_output, name='ViT')
    return model

# ==================== BUILD EFFICIENTNET ====================
print("Building EfficientNetB0 model...")

def create_efficientnet_model():
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    
    base_model = EfficientNetB0(
        include_top=False,
        weights='imagenet',
        input_tensor=inputs,
        pooling='avg'
    )
    
    # Fine-tune last 20 layers
    for layer in base_model.layers[:-20]:
        layer.trainable = False
    
    x = base_model.output
    x = layers.Dropout(0.3)(x)
    efficientnet_output = layers.Dense(256, activation='relu', name='efficientnet_features')(x)
    
    model = keras.Model(inputs=inputs, outputs=efficientnet_output, name='EfficientNetB0')
    return model

# ==================== BUILD ENSEMBLE MODEL ====================
print("Building ensemble model...")

def create_ensemble_model():
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    
    # Create both models
    vit_model = create_vit_model()
    efficientnet_model = create_efficientnet_model()
    
    # Get features from both models
    vit_features = vit_model(inputs)
    efficientnet_features = efficientnet_model(inputs)
    
    # Concatenate features
    combined = layers.Concatenate()([vit_features, efficientnet_features])
    
    # Classification head
    x = layers.Dense(512, activation='relu')(combined)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)
    
    model = keras.Model(inputs=inputs, outputs=outputs, name='Ensemble_Model')
    return model

ensemble_model = create_ensemble_model()

print(f"âœ“ Ensemble model created")
print(f"  Total parameters: {ensemble_model.count_params():,}")

# ==================== COMPILE MODEL ====================
ensemble_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("âœ“ Model compiled with Adam optimizer")

# ==================== CUSTOM CALLBACK ====================
class DetailedProgressCallback(keras.callbacks.Callback):
    def __init__(self, target_accuracy=0.95):
        super().__init__()
        self.target_accuracy = target_accuracy
        self.best_val_acc = 0
    
    def on_epoch_begin(self, epoch, logs=None):
        print(f"\n{'='*70}")
        print(f"Epoch {epoch + 1}/{EPOCHS}")
        print(f"{'='*70}")
    
    def on_epoch_end(self, epoch, logs=None):
        print(f"\n{'â”€'*70}")
        print(f"Epoch {epoch + 1} Results:")
        print(f"  Training   â†’ Loss: {logs['loss']:.4f} | Accuracy: {logs['accuracy']:.4f}")
        print(f"  Validation â†’ Loss: {logs['val_loss']:.4f} | Accuracy: {logs['val_accuracy']:.4f}")
        
        if logs['val_accuracy'] > self.best_val_acc:
            self.best_val_acc = logs['val_accuracy']
            print(f"  ğŸ¯ New best validation accuracy: {self.best_val_acc:.4f}")
        
        if logs['val_accuracy'] >= self.target_accuracy:
            print(f"  âœ“ Target accuracy {self.target_accuracy:.2%} achieved!")
        
        print(f"{'â”€'*70}")

# ==================== CALLBACKS ====================
print("\nConfiguring callbacks...")

callbacks = [
    DetailedProgressCallback(target_accuracy=TARGET_ACCURACY),
    ModelCheckpoint(
        filepath=os.path.join(CHECKPOINT_DIR, 'best_model_epoch_{epoch:02d}_acc_{val_accuracy:.4f}.h5'),
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    ),
    EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    )
]

print("âœ“ Callbacks configured")

# ==================== TRAINING ====================
print("\n" + "=" * 70)
print("STARTING TRAINING")
print("=" * 70)

steps_per_epoch = train_generator.samples // BATCH_SIZE
validation_steps = validation_generator.samples // BATCH_SIZE

history = ensemble_model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    callbacks=callbacks,
    verbose=1
)

# ==================== SAVE FINAL MODEL ====================
print("\n" + "=" * 70)
print("SAVING FINAL MODEL")
print("=" * 70)

ensemble_model.save(MODEL_SAVE_PATH)
print(f"âœ“ Model saved to: {MODEL_SAVE_PATH}")

# ==================== EVALUATION ====================
print("\n" + "=" * 70)
print("MODEL EVALUATION")
print("=" * 70)

# Get predictions
validation_generator.reset()
y_pred_probs = ensemble_model.predict(validation_generator, steps=validation_steps, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = validation_generator.classes[:len(y_pred)]

# Classification Report
class_names = list(train_generator.class_indices.keys())
print("\nClassification Report:")
print("=" * 70)
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Overall accuracy
accuracy = np.mean(y_pred == y_true)
print(f"\nFinal Test Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")

if accuracy >= TARGET_ACCURACY:
    print(f"âœ“ TARGET ACHIEVED: {accuracy:.2%} >= {TARGET_ACCURACY:.2%}")
else:
    print(f"âš  Target not reached: {accuracy:.2%} < {TARGET_ACCURACY:.2%}")

# ==================== VISUALIZATIONS ====================
print("\nGenerating visualizations...")

fig = plt.figure(figsize=(18, 5))

# Accuracy Plot
ax1 = plt.subplot(1, 3, 1)
ax1.plot(history.history['accuracy'], label='Train Accuracy', linewidth=2, marker='o')
ax1.plot(history.history['val_accuracy'], label='Val Accuracy', linewidth=2, marker='s')
ax1.axhline(y=TARGET_ACCURACY, color='r', linestyle='--', label=f'Target ({TARGET_ACCURACY:.0%})')
ax1.set_title('Model Accuracy', fontsize=14, fontweight='bold')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Loss Plot
ax2 = plt.subplot(1, 3, 2)
ax2.plot(history.history['loss'], label='Train Loss', linewidth=2, marker='o')
ax2.plot(history.history['val_loss'], label='Val Loss', linewidth=2, marker='s')
ax2.set_title('Model Loss', fontsize=14, fontweight='bold')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Confusion Matrix
ax3 = plt.subplot(1, 3, 3)
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, 
            yticklabels=class_names, ax=ax3, cbar_kws={'label': 'Count'})
ax3.set_title('Confusion Matrix', fontsize=14, fontweight='bold')
ax3.set_ylabel('True Label')
ax3.set_xlabel('Predicted Label')

plt.tight_layout()
plt.savefig('/teamspace/studios/this_studio/training_results.png', dpi=150, bbox_inches='tight')
print("âœ“ Visualization saved: training_results.png")
plt.show()

# ==================== FINAL SUMMARY ====================
print("\n" + "=" * 70)
print("TRAINING COMPLETED")
print("=" * 70)

final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
best_val_acc = max(history.history['val_accuracy'])
best_epoch = history.history['val_accuracy'].index(best_val_acc) + 1

print(f"\nFinal Metrics:")
print(f"  Training Accuracy:   {final_train_acc:.4f} ({final_train_acc*100:.2f}%)")
print(f"  Validation Accuracy: {final_val_acc:.4f} ({final_val_acc*100:.2f}%)")
print(f"  Best Val Accuracy:   {best_val_acc:.4f} at Epoch {best_epoch}")
print(f"\n  Training Loss:       {history.history['loss'][-1]:.4f}")
print(f"  Validation Loss:     {history.history['val_loss'][-1]:.4f}")

# Check for overfitting/underfitting
train_val_gap = abs(final_train_acc - final_val_acc)
if train_val_gap > 0.05:
    print(f"\nâš  Potential overfitting detected (gap: {train_val_gap:.4f})")
elif final_val_acc < 0.80:
    print(f"\nâš  Potential underfitting detected (val_acc: {final_val_acc:.4f})")
else:
    print(f"\nâœ“ Model shows good generalization (gap: {train_val_gap:.4f})")

print(f"\n{'='*70}")
print(f"âœ“ Model saved: {MODEL_SAVE_PATH}")
print(f"âœ“ Results plot: /teamspace/studios/this_studio/training_results.png")
print(f"âœ“ Checkpoints: {CHECKPOINT_DIR}")
print(f"{'='*70}")

print("\nğŸ‰ Training pipeline completed successfully!")
print("\nTo load the model later:")
print(f"model = keras.models.load_model('{MODEL_SAVE_PATH}')")
```

```python
"""
Quick Test Script - Test the trained model on holdout data
Fixed: Handles custom layers (Patches, PatchEncoder)
"""

import os
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

print("="*70)
print("TESTING TRAINED MODEL ON UNSEEN DATA")
print("="*70)

# ==================== CONFIG ====================
DATASET_PATH = '/teamspace/studios/this_studio/finalpreprocessed'
MODEL_PATH = '/teamspace/studios/this_studio/ensemble_final_model.h5'
IMG_SIZE = 224
CLASSES = ['buyinput', 'sellinput']

# ==================== DEFINE CUSTOM LAYERS ====================
class Patches(layers.Layer):
    def __init__(self, patch_size, **kwargs):
        super().__init__(**kwargs)
        self.patch_size = patch_size

    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1, self.patch_size, self.patch_size, 1],
            strides=[1, self.patch_size, self.patch_size, 1],
            rates=[1, 1, 1, 1],
            padding="VALID",
        )
        patch_dims = patches.shape[-1]
        patches = tf.reshape(patches, [batch_size, -1, patch_dims])
        return patches
    
    def get_config(self):
        config = super().get_config()
        config.update({"patch_size": self.patch_size})
        return config

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim, **kwargs):
        super().__init__(**kwargs)
        self.num_patches = num_patches
        self.projection_dim = projection_dim
        self.projection = layers.Dense(units=projection_dim)
        self.position_embedding = layers.Embedding(
            input_dim=num_patches, output_dim=projection_dim
        )

    def call(self, patch):
        positions = tf.range(start=0, limit=self.num_patches, delta=1)
        encoded = self.projection(patch) + self.position_embedding(positions)
        return encoded
    
    def get_config(self):
        config = super().get_config()
        config.update({
            "num_patches": self.num_patches,
            "projection_dim": self.projection_dim
        })
        return config

# ==================== LOAD MODEL ====================
print("\nLoading model with custom layers...")
custom_objects = {
    'Patches': Patches,
    'PatchEncoder': PatchEncoder
}

model = keras.models.load_model(MODEL_PATH, custom_objects=custom_objects)
print("âœ“ Model loaded successfully")

# ==================== PREPARE TEST DATA ====================
print("\nCollecting images...")
all_files = []
all_labels = []

for idx, class_name in enumerate(CLASSES):
    class_path = os.path.join(DATASET_PATH, class_name)
    files = [os.path.join(class_path, f) for f in os.listdir(class_path) 
             if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]
    all_files.extend(files)
    all_labels.extend([idx] * len(files))

print(f"Total images: {len(all_files)}")

# Create 15% test set that was NEVER used in training
_, test_files, _, test_labels = train_test_split(
    all_files, all_labels, test_size=0.15, random_state=999, stratify=all_labels
)

print(f"Test set: {len(test_files)} images (15% holdout)")
print(f"  Buy signals:  {sum(1 for l in test_labels if l == 0)}")
print(f"  Sell signals: {sum(1 for l in test_labels if l == 1)}")

# Load test images
print("\nLoading test images...")
test_images = []
for file_path in test_files:
    img = Image.open(file_path).convert('RGB')
    img = img.resize((IMG_SIZE, IMG_SIZE))
    test_images.append(np.array(img) / 255.0)

test_images = np.array(test_images)
test_labels = np.array(test_labels)

print(f"âœ“ Loaded {len(test_images)} images")
print(f"  Shape: {test_images.shape}")

# ==================== PREDICT ====================
print("\nRunning predictions...")
predictions_probs = model.predict(test_images, batch_size=32, verbose=0)
predictions = np.argmax(predictions_probs, axis=1)

# ==================== RESULTS ====================
print("\n" + "="*70)
print("TEST RESULTS ON COMPLETELY UNSEEN DATA")
print("="*70)

accuracy = np.mean(predictions == test_labels)
print(f"\nğŸ¯ TRUE TEST ACCURACY: {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"   (Validation accuracy was: 96.12%)")

print("\nClassification Report:")
print(classification_report(test_labels, predictions, target_names=CLASSES, digits=4))

print("\nConfusion Matrix:")
cm = confusion_matrix(test_labels, predictions)
print(f"                Predicted")
print(f"              Buy    Sell")
print(f"Actual Buy    {cm[0,0]:<6} {cm[0,1]:<6}")
print(f"      Sell    {cm[1,0]:<6} {cm[1,1]:<6}")

# Calculate per-class metrics
buy_accuracy = cm[0,0] / (cm[0,0] + cm[0,1])
sell_accuracy = cm[1,1] / (cm[1,0] + cm[1,1])
buy_precision = cm[0,0] / (cm[0,0] + cm[1,0]) if (cm[0,0] + cm[1,0]) > 0 else 0
sell_precision = cm[1,1] / (cm[0,1] + cm[1,1]) if (cm[0,1] + cm[1,1]) > 0 else 0

print(f"\nDetailed Metrics:")
print(f"  Buy Signals:")
print(f"    Recall:    {buy_accuracy:.4f} ({buy_accuracy*100:.2f}%)")
print(f"    Precision: {buy_precision:.4f} ({buy_precision*100:.2f}%)")
print(f"  Sell Signals:")
print(f"    Recall:    {sell_accuracy:.4f} ({sell_accuracy*100:.2f}%)")
print(f"    Precision: {sell_precision:.4f} ({sell_precision*100:.2f}%)")

# Compare with training performance
print(f"\n" + "="*70)
print("COMPARISON:")
print("="*70)
print(f"  Training Accuracy:   98.99%")
print(f"  Validation Accuracy: 96.12%")
print(f"  TRUE Test Accuracy:  {accuracy*100:.2f}%")
gap_train_test = 98.99 - (accuracy * 100)
gap_val_test = 96.12 - (accuracy * 100)
print(f"\n  Training-Test Gap:   {gap_train_test:.2f}%")
print(f"  Validation-Test Gap: {gap_val_test:.2f}%")

print("\n" + "="*70)
if accuracy >= 0.95:
    print("âœ… EXCELLENT! Model generalizes exceptionally well (â‰¥95%)")
    print("   Your model is production-ready! ğŸš€")
elif accuracy >= 0.92:
    print("âœ… VERY GOOD! Model performance is strong (92-95%)")
    print("   Model is ready for deployment with monitoring")
elif accuracy >= 0.90:
    print("âœ… GOOD! Model performance is solid (90-92%)")
    print("   Consider minor improvements but usable")
else:
    print("âš ï¸  Model shows overfitting (accuracy <90%)")
    print("   Recommend retraining with adjustments")
print("="*70)

print(f"\nâœ“ Testing complete!")
```
